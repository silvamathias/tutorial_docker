```python
import numpy as np
import matplotlib.pyplot as mp 
```

```python
a = np.linspace(0,4*np.pi,400)
```

```python
y1 = np.sin(a)
```

```python
y2 = np.cos(a)
```

```python
mp.plot(a,y1)
mp.plot(a,y2)
mp.show()
```

```python
fig, gx = mp.subplots(2,1)
gx[0].plot(a,y1)
gx[1].plot(a,y2)
mp.show()
```
